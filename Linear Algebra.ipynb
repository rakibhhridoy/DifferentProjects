```python
def plt_vector2d(vector2d, origin=[0, 0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
              head_width=0.2, head_length=0.3, length_includes_head=True,
              **options)
```

```python
plt_vector2d(a, color= 'red')
plt_vector2d(b, color = 'b')
plt_vector2d(c, color = 'c')
plt.axis([0,8,0,8])
plt.grid()
plt.show()
```

```python
def pltvector2d(vector2d, origin = [0,0], **options):
    return plt.arrow(x = origin[0], y = origin[1] , dx = vector2d[0], dy= vector2d[1], head_width = 0.2, head_length = 0.4, length_includes_head= True, **options)
```

```python
pltvector2d(a, color = 'g')
pltvector2d(b, color= 'b')
pltvector2d(c, color= 'r')
plt.axis([0,8,0,8])
plt.grid()
plt.show()
```

```python
a = np.array([2,4,3])
b = np.array([6,5,2])
c = np.array([9,10,12])
x_coords , y_coords , z_coords = zip(a,b,c)
```

```python
plt3d = plt.subplot(111, projection = '3d')
plt3d.scatter(x_coords, y_coords, z_coords, c = ['r', 'b', 'k'])
plt3d.set_zlim3d([0,18])
plt.show()
```

```python
def plt_vector3d(ax, vector3d, z0, **options):
    for v in vector3d:
        x,y,z = v
        ax.plot([x,x], [y,y], [z0, z], linestyle = 'dotted')

    x_coords, y_coords, z_coords = zip(*vector3d)
    ax.scatter(x_coords, y_coords, z_coords)
```

```python
plt3d = plt.subplot(111, projection = '3d')
plt3d.set_zlim([0,18])
plt_vector3d(plt3d, [a,b,c], 0, color = ['r', 'b', 'y'])
plt.grid()
plt.show()
```

#### Norm in Pure Python

```python
def norm_vector(vector):
    x = [i**2 for i in vector]
    return sum(x)**0.5
```

```python
print(f'Normalization vector of {a}\n ||a||: {norm_vector(a)}\n')
print(f'Normalization vector of {b}\n ||b||: {norm_vector(b)}\n')
print(f'Normalization vector of {c}\n ||c||: {norm_vector(c)}')
```
